```python
df = df.drop(['College_ID'], axis=1)
df.head()
```

```python
df = pd.get_dummies(df, columns=['Internship_Experience', 'Placement'], drop_first=True)
```

```python
df
```

```python
df.head()
```

```python
x = df.drop('Placement_Yes', axis=1)
y = df['Placement_Yes']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42)
}
```

```python
for name, model in models.items():
    model.fit(x_train, y_train)
    score = model.score(x_test, y_test)  # This gives R² score
    print(f"{name}: Score = {score:.4f}")
```

```python
# IQ , Prev_Sem_Result , CGPA ,Academic_Performance ,Internship_Experience ,Extra_Curricular_Score  ,Communication_Skills ,Projects_Completed
model.predict(np.array([[89, 8.6, 8.7, 0, 5.9, 8.0, 8, 3]]))  # Example input for prediction
```
